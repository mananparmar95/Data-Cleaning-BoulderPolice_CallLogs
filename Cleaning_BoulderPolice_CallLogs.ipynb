```
#It looks like records of the columns have interchanged --> Incident, ResponseDate. Make them perfect to use.
bpd['ResponseDate'], bpd['Incident_ID'] = np.where(bpd['ResponseDate']=='1', [bpd['Incident_ID'], bpd['ResponseDate']], [bpd['ResponseDate'], bpd['Incident_ID'] ])
bpd['ResponseDate'], bpd['Incident_ID'] = np.where(bpd['ResponseDate']=='2', [bpd['Incident_ID'], bpd['ResponseDate']], [bpd['ResponseDate'], bpd['Incident_ID'] ])
bpd['ResponseDate'], bpd['Incident_ID'] = np.where(bpd['ResponseDate']=='3', [bpd['Incident_ID'], bpd['ResponseDate']], [bpd['ResponseDate'], bpd['Incident_ID'] ])
bpd['ResponseDate'], bpd['Incident_ID'] = np.where(bpd['ResponseDate']=='4', [bpd['Incident_ID'], bpd['ResponseDate']], [bpd['ResponseDate'], bpd['Incident_ID'] ])
```

```
#It looks like there is an interchange of values between Porblem and CaseNumber columns too! Fix it!
"""One pattern found from this data is that, all the case numbers start with the number 2. We will use that to interchange the columns"""
```

```
#One important pattern found from this data is that, all the case numbers start with the number 2. We will use that to interchange the column
bpd['Problem'],bpd['CaseNumber'] =np.where(bpd['Problem'].astype(str).str[0]=='2', [bpd['CaseNumber'],bpd['Problem']]  ,[bpd['Problem'],bpd['CaseNumber']]  )
```

```
bpd
```

```
#Filling nullvalues in one dataframe with non-null values from the other using .combine_first()
bpd.Problem= bpd.Problem.combine_first(bpd.CaseNumber)
bpd
```

```
#Now Problem column is Perfect
bpd["Problem"].value_counts()
```

```
#Check the column Case Number. It looks incorrect
bpd["CaseNumber"].value_counts()
```

```
#Note: Here that the Case Number column has some Problems, so we will convert them as NaN
bpd['CaseNumber']= np.where(bpd['CaseNumber'].astype(str).str[0]!='2', 'NaN', bpd['CaseNumber'])
```
